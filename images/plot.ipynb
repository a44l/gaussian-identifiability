```julia
using Plots, CSV, DataFrames;
```

```julia
df_4 = CSV.read("../data/secant-dimensions-deg4.csv", DataFrame);
df_5 = CSV.read("../data/secant-dimensions-deg5.csv", DataFrame);
df_6 = CSV.read("../data/secant-dimensions-deg6.csv", DataFrame);
```

```julia
N_max = 18;
para_count_rk_4 = [binomial(n+3, 4)/(binomial(n+1, 2)+n) for n=df_5.n];
para_count_rk_5 = [binomial(n+4, 5)/(binomial(n+1, 2)+n) for n=df_5.n];
para_count_rk_6 = [binomial(n+5, 6)/(binomial(n+1, 2)+n) for n=df_6.n];
plt = plot(df_5.n[1:N_max], para_count_rk_5[1:N_max], xlabel = "Dimension (n)", xticks=min(df_5.n[1:N_max]...):2:max(df_5.n[1:N_max]...), ylabel = "Maximum nondefective rank (m)",
    title = "Max. nondef. Ranks of Gaussian Moment Variety",
    legend = :topleft, label = "degree 5", seriestype=:line, linestyle=:dash, color=1)
plot!(plt, df_5.n[1:N_max], df_5.rank[1:N_max], legend = :topleft, label = "", seriestype=:scatter, color=1)
plot!(plt, df_6.n[1:N_max], para_count_rk_6[1:N_max], legend = :topleft, label = "degree 6", seriestype=:line, linestyle=:dash, color=2)
plot!(plt, df_6.n[1:N_max], df_6.rank[1:N_max], legend = :topleft, label = "", seriestype=:scatter, color=2)


# Adding a descriptive text
# annotate!(plt, 0.2, 0.9, text("Identifiability of the Moment Variety", :top, 12))

# Displaying the plot
display(plt)
```
